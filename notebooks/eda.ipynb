# Exploratory Data Analysis (EDA)
This notebook explores the IMDB dataset of 50,000 movie reviews.  
We aim to understand the distribution of sentiments, common words in reviews, and review lengths.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
df = pd.read_csv("C:/Users/Himayaraju/OneDrive/Tài liệu/movie-sentiment-analysis/IMDB Dataset.csv")
df.head()
```

## Load Dataset
We begin by loading the IMDB dataset and inspecting the first few rows.


```python
sns.countplot(x='sentiment', data=df)
plt.title("Sentiment Distribution")
plt.show()
```

## Load Dataset
We begin by loading the IMDB dataset and inspecting the first few rows.


```python
positive_text = " ".join(df[df['sentiment']=='positive']['review'])
negative_text = " ".join(df[df['sentiment']=='negative']['review'])

wordcloud_pos = WordCloud(width=800, height=400, background_color='white').generate(positive_text)
wordcloud_neg = WordCloud(width=800, height=400, background_color='black').generate(negative_text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud_pos, interpolation='bilinear')
plt.axis("off")
plt.title("Positive Reviews Word Cloud")
plt.show()

plt.figure(figsize=(10,5))
plt.imshow(wordcloud_neg, interpolation='bilinear')
plt.axis("off")
plt.title("Negative Reviews Word Cloud")
plt.show()
```

## Word Clouds
Visualizing the most frequent words in positive and negative reviews.

```python
df['review_length'] = df['review'].apply(lambda x: len(x.split()))
sns.histplot(df['review_length'], bins=50)
plt.title("Distribution of Review Lengths")
plt.show()
```

## Word Clouds
Visualizing the most frequent words in positive and negative reviews.


```python
## Insights
- Dataset is balanced:25000 positive, 25000 negative reviews.
- Positive reviews often contain words like "great", "love", "excellent".
- Negative reviews often contain words like "bad", "boring", "worst".
- Review lengths vary widely, with most between 50–200 words.
```
